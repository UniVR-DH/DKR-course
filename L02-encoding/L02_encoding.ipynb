```python
# Taking a binary string, decode it as number, then decode it as character 

test_string = [
'01001000', # H
'01100101', # e
'01101100', # l
'01101100', # l
'01101111', # o
'00100000', # space
'01110111', # w
'01101111', # o
'01110010', # r
'01101100', # l
'01100100', # d
'00100001'  # !
]

int_values = [ int(t,2) for t in test_string ]


# chr() assumes the UTF-8 encoding
t = ''.join(chr(n)  for n in int_values)

print(t)
```

```python
# Taking a binary string, decode it as number, then decode it as character 

# Here we changed some 0 into 1 at the beginning of the bit strings

test_string = [
'11001000', # Ãˆ
'01100101', # 
'01101100', # 
'01101100', # 
'01101111', # 
'00100000', # space
'11110111', # ?
'11101111', # ?
'11110010', # ?
'11101100', # ?
'11100100', # ?
'10100001'  # ?
]

int_values = [ int(t,2) for t in test_string ]


# chr() assumes the UTF-8 encoding
t = ''.join(chr(n)  for n in int_values)

print(t)
```

```python
print("Hello!".encode("utf-8"))  # byte representation using utf-8

print("Hello!".encode("utf-16"))  # byte representation using utf-16

print("Hello RÃ©sumÃ©".encode("utf-8"))

print("Hello RÃ©sumÃ©".encode("utf-16"))

print("Hello RÃ©sumÃ©".encode("iso-8859-1")) # byte representation using iso-8859-1
print("Hello RÃ©sumÃ©".encode("latin-1")) # .. also called latin-1

print("Hello RÃ©sumÃ©".encode("gb18030")) # Unified Chinese

# for more see  https://docs.python.org/3/library/codecs.html#standard-encodings
```
